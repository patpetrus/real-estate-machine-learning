```python
import pandas as pd
```

```python
df = pd.read_excel("intrates.xlsx")
```

```python
df.head()
```

```python
df_grouped = df.groupby(["Month","Year"]).mean("InterestRate")
```

```python
df_grouped.reset_index(inplace=True)
df_grouped.head()
```

```python
df_groupedsale = df_grouped
```

```python
df_groupedsale.rename(columns={"Interest Rate": "Interest Rate 2"})
```

```python
df_2= pd.read_excel("indy tax new calcs.xlsx")

df_2.head()
```

```python
df_2[["Prior month","Prior Year"]]
```

```python
new_df= df_2.merge(df_grouped, how= 'left', left_on=["Prior month","Prior Year"], right_on =["Month","Year"])
new_df.head()
```

```python
new_df[["Prior month","Prior Year","Month","Year","Interest Rate"]]
```

```python
newest_df= new_df.merge(df_groupedsale, how= 'left', left_on=["Sale month","Sale Year"], right_on =["Month","Year"])
newest_df.head()
```

```python
newest_df.head()
```

```python
newest_df.to_excel("pandas.xlsx")
```
